```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics as stats
```

```python
data = np.linspace(0,10,10000)
```

```python
data
```

```python
len(data)
```

```python
np.mean(data)
```

```python
np.median(data)
```

```python
stats.mode(data)
```

```python
np.std(data)
```

```python
norm.pdf(data)
```

```python
plt.figure(figsize=(10,10))
plt.plot(data,norm.pdf(data))
```
